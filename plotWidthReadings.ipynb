```python
import csv
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
from operator import add
from os import listdir
from os.path import isfile, join
```

```python
files = ['widths/'+file for file in listdir('widths') if isfile(join('widths', file))]
files
```

```python
left = []
centre = []
right = []
dist = []

with open(files[0], newline='') as csvfile:
    reader = csv.reader(csvfile)

    for row in reader:
        try:
            left.append(int(row[1]))
            centre.append(int(row[2]))
            right.append(int(row[3]))
            dist.append(float(row[7]))
        except ValueError:
            pass

b,a = signal.butter(5, 0.05, btype='lowpass')

plt.plot(dist, signal.filtfilt(b,a,left))
plt.plot(dist, signal.filtfilt(b,a,centre))
plt.plot(dist, signal.filtfilt(b,a,right))

plt.grid()
plt.title(files[0])
plt.xlabel("mm")
plt.ylabel("Sensor Reading")
plt.legend(['left', 'centre', 'right'])
plt.show()
```

```python
""" Plot bumper readings for each file """

for file in files:

    lbump = []
    rbump = []
    dist = []

    with open(file, newline='') as csvfile:
        reader = csv.reader(csvfile)

        for row in reader:
            try:
                lbump.append(int(row[5]))
                rbump.append(int(row[6]))
                dist.append(float(row[7]))
            except ValueError:
                pass

    b,a = signal.butter(5, 0.05, btype='lowpass')

    bumpSum = [lbump[i] + rbump[i] for i in range(len(lbump))]

    plt.plot(dist, signal.filtfilt(b,a,bumpSum))

plt.grid()
plt.title('Plot of sum of bumper readings for each file')
plt.xlabel("mm")
plt.ylabel("Sensor Reading")
plt.legend(files)
plt.show()
```
